# JupyterLab First Look

## Installing JupyterLab:

This part was easy, once I was able to get the configuration settings from Jupyter Notebook to stop interfereing with the JupyterLab launch process. 


- pip install jupyterlab
- jupyter serverextension enable --py jupyterlab --sys-prefix
- rm -rf ~/.jupyter/
- jupyter lab

## Using JupyterLab

There are 4 different environments that can be launched in the lab:
- Notebook
- Terminal
- Text Editor
- Code Console

## JupyterLab Extensions

JupyterLab has streamlined the way that extensions can be added by providing a build script for generating third party extension JavaScript bundles.

You can install your etension with:
- jupyter labextension install path/to/my_extension my_extension [--user|--sys-prefix]
- jupyter labextension enable my_extension [--sys-prefix]

This type of design should lend itself to be very enabling for users as these features are better integrated. .
