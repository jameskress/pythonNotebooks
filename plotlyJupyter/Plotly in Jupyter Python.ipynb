```python
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
# Import data from csv
dr = pd.read_csv('restartTimings.csv', delimiter="\t")
dr.head()
```

```python
trace1 = go.Scatter(
                    x=dr['Timestep'], y=dr['Restart Max'], # Data
                    mode='lines', name='Max Restart Write Time' # Additional options
                   )
trace2 = go.Scatter(x=dr['Timestep'], y=dr['Restart Min'], mode='lines', name='Min Restart Write Time', yaxis='y2' )
#trace3 = go.Scatter(x=df['x'], y=df['cosx'], mode='lines', name='cosx')

layout = go.Layout(title='XGC Run Restart Write Timings',
                   plot_bgcolor='rgb(230, 230,230)',
                yaxis=dict(
                    title='Max Write Time (sec)'
                ),
                yaxis2=dict(
                    title='Min Write Time (sec)',
                    titlefont=dict(
                        color='rgb(148, 103, 189)'
                    ),
                    tickfont=dict(
                        color='rgb(148, 103, 189)'
                    ),
                    overlaying='y',
                    side='right'
                ))

#fig = go.Figure(data=[trace1, trace2, trace3], layout=layout)
fig = go.Figure(data=[trace1, trace2], layout=layout)

# Plot data in the notebook
py.iplot(fig, filename='simple-plot-from-csv')
```

```python
# Import data from csv
df = pd.read_csv('isosurfaceTimes.csv', delimiter="\t")
df.head()
```

```python
trace1 = go.Scatter(
                    x=df['ts'], y=df['iso1'], # Data
                    mode='lines+markers', name='iso1 Time' # Additional options
                   )
trace2 = go.Scatter(
                    x=df['ts'], y=df['iso2'], # Data
                    mode='lines+markers', name='iso2 Time' # Additional options
                   )
trace3 = go.Scatter(
                    x=df['ts'], y=df['iso3'], # Data
                    mode='lines+markers', name='iso3 Time' # Additional options
                   )
trace4 = go.Scatter(
                    x=df['ts'], y=df['iso4'], # Data
                    mode='lines+markers', name='iso4 Time' # Additional options
                   )
trace5 = go.Scatter(
                    x=df['ts'], y=df['iso4'], # Data
                    mode='lines+markers', name='iso5 Time' # Additional options
                   )
trace6 = go.Bar(x=df['ts'], 
                y=df['dataTransfered'], 
                name='Data Transfered', 
                yaxis='y2', 
                opacity=.5,
                marker=dict(
                    color='rgb(158,202,225)',
                    line=dict(
                        color='rgb(8,48,107)',
                        width=1.5),
                ))

layout = go.Layout(title='XGC Run Restart Write Timings',
                   plot_bgcolor='rgb(230, 230,230)',
                legend=dict(
                    x=1.05,
                    y=1
                ),
                yaxis=dict(
                    title='Time (sec)'
                ),
                yaxis2=dict(
                    title='MB',
                    titlefont=dict(
                        color='rgb(148, 103, 189)'
                    ),
                    tickfont=dict(
                        color='rgb(148, 103, 189)'
                    ),
                    overlaying='y',
                    side='right'
                ),
                xaxis=dict(
                    title='Timestep'
                ))

fig = go.Figure(data=[trace1, trace2, trace3, trace4, trace5, trace6], layout=layout)

# Plot data in the notebook
py.iplot(fig, filename='simple-plot-from-csv')
```
